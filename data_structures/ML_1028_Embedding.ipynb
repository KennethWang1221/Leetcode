## ML_1028: Embedding Layer

**Difficulty**: Easy | **TorchCode**: #18

### Problem
Implement an embedding lookup table as an `nn.Module`. Store a learnable weight matrix and return rows indexed by input indices. Supports batched index tensors of any shape.

### Formula
$$\text{Embedding}(i) = W[i], \quad W \in \mathbb{R}^{N_{\text{vocab}} \times d_{\text{embed}}}$$

```python
import torch
import torch.nn as nn

class MyEmbedding(nn.Module):
    def __init__(self, num_embeddings, embedding_dim):
        super().__init__()
        self.weight = nn.Parameter(torch.randn(num_embeddings, embedding_dim))

    def forward(self, indices):
        return self.weight[indices]
```

```python
import torch
import torch.nn as nn

# --- Test 1: Weight shape ---
emb = MyEmbedding(100, 32)
assert isinstance(emb, nn.Module)
assert emb.weight.shape == (100, 32)
assert emb.weight.requires_grad

# --- Test 2: Lookup correctness ---
emb = MyEmbedding(10, 4)
idx = torch.tensor([0, 3, 7])
out = emb(idx)
assert out.shape == (3, 4)
assert torch.equal(out[0], emb.weight[0])
assert torch.equal(out[1], emb.weight[3])

# --- Test 3: Batch of indices ---
emb = MyEmbedding(20, 8)
idx = torch.tensor([[1, 2], [3, 4]])
assert emb(idx).shape == (2, 2, 8)

# --- Test 4: Gradient flow ---
emb = MyEmbedding(10, 4)
emb(torch.tensor([2, 5])).sum().backward()
assert emb.weight.grad is not None
assert emb.weight.grad[2].abs().sum() > 0
assert emb.weight.grad[0].abs().sum() == 0

print('All tests passed!')
```
