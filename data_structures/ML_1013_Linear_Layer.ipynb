## ML_1013: Simple Linear Layer

**Difficulty**: Medium | **TorchCode**: #03

### Problem
Implement a fully-connected linear layer from scratch (no `nn.Linear`). Initialize weights with Kaiming scaling and support gradient computation.

### Formula
$$y = xW^T + b, \quad W \sim \mathcal{N}\!\left(0,\, \frac{1}{\sqrt{d_{\text{in}}}}\right)$$

```python
import torch
import math

class SimpleLinear:
    def __init__(self, in_features: int, out_features: int):
        self.weight = torch.randn(out_features, in_features) * (1 / math.sqrt(in_features))
        self.weight.requires_grad_(True)
        self.bias = torch.zeros(out_features, requires_grad=True)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return x @ self.weight.T + self.bias
```

```python
import torch

# --- Test 1: Weight & bias shapes ---
layer = SimpleLinear(8, 4)
assert layer.weight.shape == (4, 8)
assert layer.bias.shape == (4,)
assert layer.weight.requires_grad
assert layer.bias.requires_grad

# --- Test 2: Forward pass correctness ---
x = torch.randn(2, 8)
y = layer.forward(x)
assert y.shape == (2, 4)
assert torch.allclose(y, x @ layer.weight.T + layer.bias, atol=1e-5)

# --- Test 3: Gradient flow ---
y = layer.forward(x)
y.sum().backward()
assert layer.weight.grad is not None
assert layer.bias.grad is not None

print('All tests passed!')
```
