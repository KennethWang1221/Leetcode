## ML_1029: GELU Activation

**Difficulty**: Easy | **TorchCode**: #19

### Problem
Implement the GELU (Gaussian Error Linear Unit) activation function using the exact formula with the error function. Used in BERT, GPT, and most modern transformers.

### Formula
$$\text{GELU}(x) = x \cdot \frac{1}{2}\left(1 + \text{erf}\!\left(\frac{x}{\sqrt{2}}\right)\right)$$

```python
import torch
import math

def my_gelu(x):
    return 0.5 * x * (1.0 + torch.erf(x / math.sqrt(2.0)))
```

```python
import torch

# --- Test 1: Matches F.gelu ---
torch.manual_seed(0)
x = torch.randn(4, 8)
assert torch.allclose(my_gelu(x), torch.nn.functional.gelu(x), atol=1e-4)

# --- Test 2: gelu(0) = 0 ---
assert torch.allclose(my_gelu(torch.tensor([0.0])), torch.tensor([0.0]), atol=1e-7)

# --- Test 3: Shape preservation ---
x = torch.randn(2, 3, 4)
assert my_gelu(x).shape == x.shape

# --- Test 4: Gradient flow ---
x = torch.randn(4, 8, requires_grad=True)
my_gelu(x).sum().backward()
assert x.grad is not None and x.grad.shape == x.shape

print('All tests passed!')
```
