## ML_1012: Implement Softmax

**Difficulty**: Easy | **TorchCode**: #02

### Problem
Implement numerically-stable Softmax from scratch without using `torch.softmax`. Subtract the max before exponentiating to avoid overflow.

### Formula
$$\text{softmax}(x_i) = \frac{e^{x_i - \max(x)}}{\sum_j e^{x_j - \max(x)}}$$

```python
import torch

def my_softmax(x: torch.Tensor, dim: int = -1) -> torch.Tensor:
    x_max = x.max(dim=dim, keepdim=True).values
    e_x = torch.exp(x - x_max)
    return e_x / e_x.sum(dim=dim, keepdim=True)
```

```python
import torch

# --- Test 1: Basic 1-D ---
x = torch.tensor([1.0, 2.0, 3.0])
out = my_softmax(x, dim=-1)
assert torch.allclose(out, torch.softmax(x, dim=-1), atol=1e-5)

# --- Test 2: 2-D along dim=-1 ---
x = torch.randn(4, 8)
out = my_softmax(x, dim=-1)
assert out.shape == x.shape
assert torch.allclose(out, torch.softmax(x, dim=-1), atol=1e-5)
assert torch.allclose(out.sum(dim=-1), torch.ones(4), atol=1e-5)

# --- Test 3: Numerical stability with large values ---
x = torch.tensor([1000., 1001., 1002.])
out = my_softmax(x, dim=-1)
assert not torch.isnan(out).any()
assert not torch.isinf(out).any()
assert torch.allclose(out, torch.softmax(x, dim=-1), atol=1e-5)

print('All tests passed!')
```
