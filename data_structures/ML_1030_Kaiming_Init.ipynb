## ML_1030: Kaiming Initialization

**Difficulty**: Easy | **TorchCode**: #20

### Problem
Implement Kaiming (He) weight initialization for ReLU networks. Fill the weight tensor in-place with samples from a normal distribution scaled to preserve variance through ReLU layers.

### Formula
$$W \sim \mathcal{N}\!\left(0,\, \sqrt{\frac{2}{n_{\text{in}}}}\right), \quad n_{\text{in}} = W.\text{shape}[1]$$

```python
import torch
import math

def kaiming_init(weight):
    fan_in = weight.shape[1] if weight.dim() >= 2 else weight.shape[0]
    std = math.sqrt(2.0 / fan_in)
    with torch.no_grad():
        weight.normal_(0, std)
    return weight
```

```python
import torch
import math

# --- Test 1: Mean approximately 0 ---
torch.manual_seed(0)
w = torch.empty(256, 512)
kaiming_init(w)
assert abs(w.mean().item()) < 0.02

# --- Test 2: Std matches sqrt(2/fan_in) ---
torch.manual_seed(0)
fan_in = 1024
w = torch.empty(256, fan_in)
kaiming_init(w)
expected = math.sqrt(2.0 / fan_in)
assert abs(w.std().item() - expected) < 0.005

# --- Test 3: Returns same tensor (in-place) ---
w = torch.empty(64, 32)
out = kaiming_init(w)
assert out is w
assert out.shape == (64, 32)

# --- Test 4: Smaller fan_in gives larger std ---
w1 = torch.empty(64, 16); w2 = torch.empty(64, 256)
kaiming_init(w1); kaiming_init(w2)
assert w1.std().item() > w2.std().item()

print('All tests passed!')
```
