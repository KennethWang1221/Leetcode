```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None


def make_list(arr):
    head_node = None
    p_node = None
    for a in arr:
        new_node = ListNode(a)
        if head_node is None:
            head_node = new_node
            p_node = new_node
        else:
            p_node.next = new_node
            p_node = new_node
    return head_node


def print_list(head):
    while head is not None:
        print(head.val, end=',')
        head = head.next
    print()

#[4, 2, 1, 3]

class Solution:
    def sortList(self, head):
        """
        1. This algorithm can not use insert_sort() stragety, I don't know why. I should think about that
        2. We use divide and conquer strategy. 
        3. First we use slow/fast pointer to split the original linkedlist to seperate linkedlist
        4. we use Merge Two Linkedlist idea to merge the splited linkedlist
        5. This idea is similar like merge_sort() in array sorting problem
        """

        dummy = ListNode(0)
        dummy.next = head 
        pre = dummy

        
        slow = head
        fast = head # define slow and fast pointer
        """
        this statement is like 
        if len(array) <=1:
            return array
        """
        if head ==None or head.next ==None: 
            return head
        

        while fast != None and fast.next !=None and fast.next.next !=None:
            slow = slow.next
            fast = fast.next.next # i++/ j++
        
        mid = slow.next # use mid as the right head
        slow.next = None # we must set slow.next =None, because this can end of recurisve process

        left= self.sortList(head) # Important, the head of left is not slow !! it is head!!!
        right = self.sortList(mid)
        """
        it is same as 
        left = array[:middle]
        right =array[middle:]
        """
        return self.merge_sort(left,right)
    
    def merge_sort(self,left,right):
        """
        we use the same idea of 21.Merge Two LinkedList 
        """
        dummy = ListNode(0)
        #dummy.next = head
        tail = dummy

        while left !=None and right !=None:
            if left.val < right.val:
                tail.next = ListNode(left.val)
                tail = tail.next
                left = left.next
            else:
                tail.next = ListNode(right.val)
                tail = tail.next
                right = right.next
        while left !=None:
            tail.next = ListNode(left.val)
            tail = tail.next
            left= left.next
        while right !=None:
            tail.next = ListNode(right.val)
            tail = tail.next
            right= right.next
        return dummy.next

head = make_list([4,2,1,3])
print_list(head)
s = Solution()
head = s.sortList(head)
print_list(head)
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def sortList(self, head: 'Optional[ListNode]') -> 'Optional[ListNode]':
        if not head:
            return head

        # Convert linked list to array
        current = head
        values = []
        while current:
            values.append(current.val)
            current = current.next

        # Sort the array
        values.sort()

        # Convert array back to sorted linked list
        new_head = ListNode(0)
        new = new_head
        for value in values:
            new.next = ListNode(value)
            new = new.next

        return new_head.next

# Test cases
def create_linked_list(values):
    head = ListNode(0)
    current = head
    for value in values:
        current.next = ListNode(value)
        current = current.next
    return head.next

def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Test case 1
input_list = create_linked_list([4, 2, 1, 3])
solution = Solution()
sorted_head = solution.sortList(input_list)
print("Test case 1:")
print_linked_list(sorted_head)

# Test case 2
input_list = create_linked_list([5, 3, 8, 7, 2])
solution = Solution()
sorted_head = solution.sortList(input_list)
print("Test case 2:")
print_linked_list(sorted_head)

# Test case 3
input_list = create_linked_list([])
solution = Solution()
sorted_head = solution.sortList(input_list)
print("Test case 3:")
print_linked_list(sorted_head)
```
