## Method1 - Divide-Conquer - Recursive
https://www.youtube.com/watch?v=q5a5OiGbT6Q

```python
from typing import List

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        if not lists:
            return None

        while len(lists) > 1:
            mergedLists = []
            for i in range(0, len(lists), 2):
                l1 = lists[i]
                l2 = lists[i + 1] if (i + 1) < len(lists) else None
                mergedLists.append(self.mergeTwoLists(l1, l2))
            lists = mergedLists
        return lists[0]

    def mergeTwoLists(self, l1, l2):
        dummy = ListNode()
        tail = dummy

        while l1 and l2:
            if l1.val < l2.val:
                tail.next = l1
                l1 = l1.next
            else:
                tail.next = l2
                l2 = l2.next
            tail = tail.next
        tail.next = l1 if l1 else l2
        return dummy.next

# Test cases
def list_to_array(head: ListNode) -> list:
    array = []
    while head:
        array.append(head.val)
        head = head.next
    return array

def array_to_list(array: list) -> ListNode:
    dummy = ListNode()
    current = dummy
    for val in array:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

# Test case 1: Normal case with multiple lists
lists = [
    array_to_list([1, 4, 5]),
    array_to_list([1, 3, 4]),
    array_to_list([2, 6])
]
solution = Solution()
merged_head = solution.mergeKLists(lists)
print(list_to_array(merged_head))  # Expected output: [1, 1, 2, 3, 4, 4, 5, 6]

# Test case 2: Empty list of lists
lists = []
merged_head = solution.mergeKLists(lists)
print(list_to_array(merged_head))  # Expected output: []

# Test case 3: List of empty lists
lists = [None, None, None]
merged_head = solution.mergeKLists(lists)
print(list_to_array(merged_head))  # Expected output: []

# Test case 4: Single list
lists = [array_to_list([1, 2, 3])]
merged_head = solution.mergeKLists(lists)
print(list_to_array(merged_head))  # Expected output: [1, 2, 3]
```
