## ML_1018: Implement RMSNorm

**Difficulty**: Medium | **TorchCode**: #08

### Problem
Implement Root Mean Square Layer Normalization. Unlike LayerNorm, RMSNorm skips mean subtraction — only normalizes by the RMS, then scales by a learnable weight. Used in LLaMA, Mistral, Qwen.

### Formula
$$\text{RMSNorm}(x) = \frac{x}{\sqrt{\text{mean}(x^2) + \varepsilon}} \cdot \gamma$$

```python
import torch

def rms_norm(x, weight, eps=1e-6):
    rms = torch.sqrt(x.pow(2).mean(dim=-1, keepdim=True) + eps)
    return x / rms * weight
```

```python
import torch

# --- Test 1: Basic behavior ---
x = torch.randn(2, 8)
weight = torch.ones(8)
out = rms_norm(x, weight)
assert out.shape == x.shape
rms = torch.sqrt(x.pow(2).mean(dim=-1, keepdim=True) + 1e-6)
assert torch.allclose(out, x / rms * weight, atol=1e-5)

# --- Test 2: With learned weight ---
x = torch.randn(4, 16)
weight = torch.randn(16)
out = rms_norm(x, weight)
rms = torch.sqrt(x.pow(2).mean(dim=-1, keepdim=True) + 1e-6)
assert torch.allclose(out, x / rms * weight, atol=1e-5)

# --- Test 3: 3-D input ---
x = torch.randn(2, 4, 32)
weight = torch.ones(32)
out = rms_norm(x, weight)
assert out.shape == x.shape
rms = torch.sqrt(x.pow(2).mean(dim=-1, keepdim=True) + 1e-6)
assert torch.allclose(out, x / rms * weight, atol=1e-5)

# --- Test 4: Gradient flow ---
x = torch.randn(2, 8, requires_grad=True)
weight = torch.ones(8, requires_grad=True)
rms_norm(x, weight).sum().backward()
assert x.grad is not None and weight.grad is not None

print('All tests passed!')
```
