```python
class Node:
    def __init__(self, val: int):
        self.val = val
        self.next = None


class MyCircularQueue:

    def __init__(self, k: int):
        self.head = self.tail = None
        self.capacity = k
        self.size = 0


    def enQueue(self, value: int) -> bool:
        if self.isFull():
            return False

        node = Node(value)
        if self.size == 0:
            self.head = self.tail = node
        else:
            self.tail.next = self.tail = node

        self.size += 1

        return True


    def deQueue(self) -> bool:
        if self.isEmpty():
            return False

        self.head = self.head.next
        self.size -= 1

        return True


    def Front(self) -> int:
        return -1 if self.isEmpty() else self.head.val


    def Rear(self) -> int:
        return -1 if self.isEmpty() else self.tail.val


    def isEmpty(self) -> bool:
        return self.size == 0

    def isFull(self) -> bool:
        return self.capacity == self.size
# Test case for MyCircularQueue

# Initialize the circular queue with capacity 3
queue = MyCircularQueue(3)

# Enqueue elements
print(queue.enQueue(1))  # Expected: True
print(queue.enQueue(2))  # Expected: True
print(queue.enQueue(3))  # Expected: True
print(queue.enQueue(4))  # Expected: False, because the queue is full

# Check the front and rear elements
print(queue.Front())  # Expected: 1
print(queue.Rear())   # Expected: 3

# Dequeue one element
print(queue.deQueue())  # Expected: True

# Check the front and rear elements again
print(queue.Front())  # Expected: 2
print(queue.Rear())   # Expected: 3

# Enqueue another element
print(queue.enQueue(4))  # Expected: True

# Check the front and rear elements again
print(queue.Front())  # Expected: 2
print(queue.Rear())   # Expected: 4

# Dequeue all elements
print(queue.deQueue())  # Expected: True
print(queue.deQueue())  # Expected: True
print(queue.deQueue())  # Expected: True
print(queue.deQueue())  # Expected: False, because the queue is empty
```
