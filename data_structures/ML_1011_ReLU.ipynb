## ML_1011: Implement ReLU

**Difficulty**: Easy | **TorchCode**: #01

### Problem
Implement the ReLU (Rectified Linear Unit) activation function from scratch without using `torch.relu` or `torch.clamp`.

### Formula
$$\text{ReLU}(x) = \max(0, x)$$

```python
import torch

def relu(x: torch.Tensor) -> torch.Tensor:
    return x * (x > 0).float()
```

```python
import torch
import time

# --- Test 1: Basic values ---
x = torch.tensor([-2., -1., 0., 1., 2.])
out = relu(x)
expected = torch.tensor([0., 0., 0., 1., 2.])
assert out.shape == expected.shape
assert torch.allclose(out, expected)

# --- Test 2: 2-D tensor ---
x = torch.randn(4, 8)
out = relu(x)
assert out.shape == x.shape
assert (out >= 0).all()
assert torch.allclose(out, x.clamp(min=0))

# --- Test 3: Gradient check ---
x = torch.tensor([-1., 0., 1., 2.], requires_grad=True)
relu(x).sum().backward()
assert x.grad is not None
assert x.grad[0] == 0.
assert x.grad[2] == 1.
assert x.grad[3] == 1.
assert x.grad[1] in (0., 1.)

# --- Test 4: Performance ---
big = torch.randn(1024, 1024)
t0 = time.perf_counter()
for _ in range(100):
    relu(big)
assert time.perf_counter() - t0 < 5.0

print('All tests passed!')
```
