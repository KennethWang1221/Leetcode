```python
# https://github.com/425776024/easy_leetcode/blob/master/easyleetcode/leetcodes/Leetcode_021_Merge_Two_Sorted_Lists.py
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None


def make_list(arr):
    head_node = None
    p_node = None
    for a in arr:
        new_node = ListNode(a)
        if head_node is None:
            head_node = new_node
            p_node = new_node
        else:
            p_node.next = new_node
            p_node = new_node
    return head_node


def print_list(head):
    while head is not None:
        print(head.val, end=',')
        head = head.next
    print()


class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        1. The important thing is how to create the new merged LinkedList
        2. And how to simulate i++/j++ in Linkedlist problem 
        3. how to give the node to the merged LinkedList
        4. this algorithm, we learn the new way to create merged LinkedList and use Node = Node.next to simulate i++
        5. This merged two linked list problem is similar with arry merge list problem
        5. return dummy.next, because we want to get the whole linkedlist
        """
        """
        dummy = ListNode(0) # this statement is so improtant. It used to create the new Linked List
        tail = dummy  # this statement used to go ahead to deeper of linkedlist
        """
        dummy = ListNode(0) # create the new merged linkedlist, and let dummy becomes the leader
        tail = dummy # dummy is used to locate the locatoion, so we use new created tail node to go head

        while l1 != None and l2 !=None: # This while loop equals to 'for i in range(len(array-1))' in array problems. We want to constriant the range, but linkedlist do not have index so we use node.next !=None as the conditional. this conditional statement indicates that one of them is at the None posiiton, but another is still indicate the useful node. but we cannot compare useful node with NONE. So we should end of this while and go to the next while
            if l1.val < l2.val: # the same as array merge list problem. 
                tail.next = ListNode(l1.val) # this line do 3 things. 1. create new ListNode in the merged list 2. we assign the l1.val to the just created new ListNode. 3. we use tail.next to point this new code, which means that we can let tail to connect new node so we do this operation again and again, we can get the final linkedlist
                tail = tail.next # this code is so important, it equals to i++/j++
                l1 = l1.next # the same, it equals to j++/i++
            else:
                tail.next = ListNode(l2.val)
                tail = tail.next
                l2 = l2.next # if l1 >= l2.val, we do this else block. It is same as array merge_sorted()!!!
        while l1!=None: # l1 !=None, this means that l2 go to NONE, but l1 still have remain nodes. Thus, we should link all the reamin node to the new merged list
            tail.next=ListNode(l1.val)
            tail = tail.next
            l1 = l1.next # this operation is like result.extend([left[i:]]) in array merge list problem
        while l2 !=None: # l2 !=None, this means that l1 go to NONE, but l2 still have remain nodes. Thus, we should link all the reamin node to the new merged list
            tail.next= ListNode(l2.val)
            tail = tail.next
            l2 = l2.next # it is like result.exten(right[j:]) ...
        return dummy.next # we want to get the whole merged linkedlist, so we just return dummy.next which can give us the whole order.
        

s = Solution()

l1 = make_list([1, 2, 4])
print_list(l1)
l2 = make_list([1,3,4])
print_list(l2)

res=s.mergeTwoLists(l1,l2)
print_list(res)
```
