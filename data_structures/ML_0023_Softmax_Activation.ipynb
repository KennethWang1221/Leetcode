```python
import math

def softmax(x):
    # Compute the exponentials of all elements
    exp_values = [math.exp(i) for i in x]
    
    # Compute the sum of the exponentials
    sum_exp_values = sum(exp_values)
    
    # Compute the probabilities by dividing each exponential by the sum
    softmax_values = [value / sum_exp_values for value in exp_values]
    
    return softmax_values

print(softmax([1, 2, 3]))
```

```python
import torch

def softmax(x, dim=-1):
    x_stable = x - torch.max(x, dim=dim, keepdim=True).values
    exp_value = torch.exp(x_stable)
    sum_exp_value = torch.sum(exp_value, dim=dim, keepdim=True)
    return exp_value / sum_exp_value

x = torch.tensor([
    [1., 2., 3.],
    [4., 5., 6.]
])

print(softmax(x, dim=-1))
print(torch.softmax(x, dim=-1))
```

```python
import math

def sigmoid(x: float) -> float:
    return 1 / (1 + math.exp(-x))

# Example usage
input_value = 2.0
print(sigmoid(input_value))  # Output: 0.8808
```
